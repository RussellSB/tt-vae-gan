# Data Exploration

A utility to easily load audio and visually debug were necessary

```python
def load_audio(path, show=False):
    # sampling rate at 16000Hz wrt flickr_audio
    x1, sr1 = librosa.load(os.path.join(path), sr=16000) 

    if(show):
        ipd.display(ipd.Audio(os.path.join(path)))
        
        t1 = np.arange(x1.shape[0]) / sr1
        plt.figure(figsize=(10, 2))
        plt.plot(t1, x1, color='gray')
        plt.xlabel('Time (seconds)')
        plt.ylabel('Amplitude')
        plt.xlim([t1[0], t1[-1]])
        plt.tight_layout()
        
    return x1, sr1
```

# Preprocessing

```python
def get_audio_titles(speaker_id):
    titles = []    
    for a in speaker_annotations:
        if a[1] == speaker_id:
            titles.append(a[0])
    return titles

titles_speaker_7 = get_audio_titles(7)
titles_speaker_4 = get_audio_titles(4)
```

```python
# n_fft=800 (0.05s), hop_length=200 (1/4 * n_fft), following Albadawy 2020
def melspec_set(titles):
    melset = []
    for title in tqdm(titles):
        x, sr = load_audio(data+title.decode())
        melset.append(librosa.feature.melspectrogram(x, sr=16000, n_fft=800, hop_length=200))
    return melset
```

```python
melset_speaker_7 = melspec_set(titles_speaker_7)
melset_speaker_4 = melspec_set(titles_speaker_4)
```

Need to get the time axis to the same length to pass spectrograms of the same dimensions through the GAN. Do this by subsampling from each melspectrogram where possible.

```python
import math
    
def melset_equalise(melset, frames):
    subsamples = []
    
    for i, m in enumerate(melset):
        frames_curr = m.shape[1] 
        n_subsamples = frames_curr / frames
        
        if n_subsamples >= 1:
            j = 0
            for _ in range(math.floor(n_subsamples)):
                m_crop = m[:, j:j+frames]
                subsamples.append(m_crop)
                j += frames

    return subsamples
```

```python
# Albadawy 2020 did a crop of 1.6s (128 frames)         
melset_7_128 = melset_equalise(melset_speaker_7, 128)    
melset_4_128 = melset_equalise(melset_speaker_4, 128)
```

A utility to save and load files with ease

```python
import pickle

def save_pickle(obj, path):
    with open(path, 'wb') as fp:
        pickle.dump(obj, fp)
        
def load_pickle(path):
    with open(path, 'rb') as fp:
        return pickle.load(fp)
```

A utility to display melspectrogram

```python
def show_mel(mel):
    plt.figure(figsize=(10, 2))
    S_dB = librosa.power_to_db(mel, ref=np.max)
    img = librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=16000, hop_length=200)
    plt.colorbar(format="%+2.f dB")
    plt.title('Mel-frequency spectrogram')
```

We note that since the male has a deeper voice (first spectrogram) the frequencies in the melspectrogram are more pressed about and below 512 Hz. The female meanwhile has her frequencies much more spread, with more spectral energy above 2048Hz and less dark empty spaces. Interestingly, the female speaker here doesn't have the best audio quality, where the microphone sometimes clips with an affect (probably from her talking too directly in to the mic). This is reflected in her audio with thick lines.

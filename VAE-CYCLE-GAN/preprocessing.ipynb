```python
def load_audio(path, show=False):
    # sampling rate at 16000Hz wrt flickr_audio
    x1, sr1 = librosa.load(os.path.join(path), sr=16000) 

    if(show):
        ipd.display(ipd.Audio(os.path.join(path)))
        
        t1 = np.arange(x1.shape[0]) / sr1
        plt.figure(figsize=(10, 2))
        plt.plot(t1, x1, color='gray')
        plt.xlabel('Time (seconds)')
        plt.ylabel('Amplitude')
        plt.xlim([t1[0], t1[-1]])
        plt.tight_layout()
        
    return x1, sr1
```

# Preprocessing

```python
def get_audio_titles(speaker_id):
    titles = []    
    for a in speaker_annotations:
        if a[1] == speaker_id:
            titles.append(a[0])
    return titles

titles_speaker_7 = get_audio_titles(7)
titles_speaker_4 = get_audio_titles(4)
```

```python
# n_fft=800 (0.05s), hop_length=200 (1/4 * n_fft), following Albadawy 2020
def melspec_set(titles):
    melset = []
    for title in tqdm(titles):
        x, sr = load_audio(data+title.decode())
        melset.append(librosa.feature.melspectrogram(x, sr=16000, n_fft=800, hop_length=200))
    return melset
```

```python
melset_speaker_7 = melspec_set(titles_speaker_7)
melset_speaker_4 = melspec_set(titles_speaker_4)
```

Need to get the time axis to the same length to pass spectrograms of the same dimensions through the GAN. Do this by subsampling from each melspectrogram where possible. This approach makes it so that if the sample is large enough it splits it and appends it to different samples that amount to the specified number of frames

```python
import math
    
def melset_equalise(melset, frames):
    subsamples = []
    
    for i, m in enumerate(melset):
        frames_curr = m.shape[1] 
        n_subsamples = frames_curr / frames
        
        if n_subsamples >= 1:
            j = 0
            for _ in range(math.floor(n_subsamples)):
                m_crop = m[:, j:j+frames]
                subsamples.append(m_crop)
                j += frames

    return subsamples
```

```python
# Albadawy 2020 did a crop of 1.6s (128 frames)         
melset_7_128 = melset_equalise(melset_speaker_7, 128)    
melset_4_128 = melset_equalise(melset_speaker_4, 128)
```

## Preprocessing (Second Approach)

In the above spectrograms were equalised on the time axis. They were equalised by cropping and then appending following files to fill in the gaps (if there were too little). Loss for generator was converging at about 25-50 at best. Going to test equalising it without appending to see if it improves results.

They crop consequetive frames, thats not what I did above. Going to set it so it is also randomly within the right range.

```python
import math
import random
    
def melset_equalise_cont(melset, frames):
    subsamples = []
    
    for i, m in enumerate(melset):
        frames_curr = m.shape[1] 
        
        # Randomly start with the crop such that the
        # Next few frames can expand to number of specified frames
        
        if frames_curr >= frames:
            start = random.randint(0, frames_curr-frames)            
            m_crop = m[:, start:start+frames]
            subsamples.append(m_crop)
        
    return subsamples
```

```python
# Albadawy 2020 did a CONTINUOUS crop of 1.6s (128 frames)         
melset_7_128_cont = melset_equalise_cont(melset_speaker_7, 128)    
melset_4_128_cont = melset_equalise_cont(melset_speaker_4, 128)
```

## Preprocessing (Third Approach)

Limiting it to just consequetive crops did diminish the dataset pair size by half. Going to try cater for this by adding some permutations for each sample. So each sample wouldnt just be paired with another one, but another two or three.

As of right now there are only repeated samples from speaker set A and not B. This may lead to slight overfitting for set A over B, but for now we will leave it as it is. The rationale for it being ok is that the pairing is atleast always unique. Repititions for samples in set B were not ensued for simplicity's sake.

```python
melset_7_128_cont = load_pickle('../pool/melset_7_128_cont.pickle')
melset_4_128_cont = load_pickle('../pool/melset_4_128_cont.pickle')
```

```python
import math
import random

# Max number of permutations per sample
# Note this assumes that melset A and B are inline with titles file (ie will not work for first approach, only second)
def melset_permute(melsetA, melsetB, maxperm):
    
    # The permutations across two lists
    permA = []
    permB = []
    
    for i, m_A in enumerate(melsetA):
        
        indexes_pairedB = [10000]
        j = 10000 # just initialising to unreachable number
        for _ in range(maxperm):
            
            # This just ensures that all pairing are unique, despite sample m_A being repeated
            while j in indexes_pairedB:
                j = random.randint(0, len(melsetB) - 1) 
                
            m_B = melsetB[j]
                
            permA.append(m_A)
            permB.append(m_B)
            
    return permA, permB
```

```python
maxperm = 2
melset_7_128_cont_p, melset_4_128_cont_p = melset_permute(melset_7_128_cont, melset_4_128_cont, maxperm)
```

## Preprocessing (Full approach)

Using mel cont splitting, but also have to save corresponding audio files for training the vocoder

```python
import random

# Returns melspectogram cropped and corresponding sample cropped
def preprocess_set(titles, frames):
    melset = []
    audioset = []
    
    for title in tqdm(titles):
        x, sr = load_audio(data+title.decode())
        m = librosa.feature.melspectrogram(x, sr=16000, n_fft=800, hop_length=200)
                                           
        frames_curr = m.shape[1] 
        
        # Randomly start with the crop such that the
        # Next few frames can expand to number of specified frames
        if frames_curr >= frames:
                                           
            # Mel frames
            start = random.randint(0, frames_curr-frames)            
            m_crop = m[:, start:start+frames]
            melset.append(m_crop)
                                           
            # Audio samples
            start = start * 200 # times[i] = frames[i] * hop_length
            length = frames * 200
            x_crop = x[start:start+length]
            audioset.append(x_crop)
        
    return melset, audioset
```

```python
melset_7_cont, audioset_7_cont = preprocess_set(titles_speaker_7, 128)
```

```python
melset_4_cont, audioset_4_cont = preprocess_set(titles_speaker_4, 128)
```
